```python
import numpy as np
import pandas as pd
```

## 產生隨機資料

```python
data = np.random.randint(30, 101, (10, 4))
```

## 轉換成 float

```python
data = data.astype(float)
```

```python
for i in range(5):
  x = np.random.randint(0, 10)
  y = np.random.randint(0, 4)
  data[x, y] = np.NaN
```

```python
data
```

## 轉換成 DataFrame

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.columns = ['國文','英文','數學','理化']
```

```python
df
```

```python
df.index = ["林怡婷","趙必桂","錢詩婷","鄭泰州","陳祐侑","林惠貴","藍以舜","韓清勇","黃子喬","張揚芃"]
```

```python
df
```

```python
df.info()
```

## 刪除缺失值

* dropna()

```python
df.dropna()
```

```python
df.dropna(subset=['國文'])
```

```python
df['國文'].notnull()
```

```python
df[df['國文'].notnull()]
```

```python
df.dropna(axis=1)
```

## 填補缺失值

* fillna()

```python
df['國文'].fillna(value=df['國文'].mean())
```

```python
df
```

```python
df.fillna(method='backfill')
```

## 簡單資料分析

```python
df = df.fillna(method='backfill')
```

```python
df
```

```python
df['國文'] < 60
```

```python
df[df['國文'] < 60]
```

```python
df[(df['國文'] < 60) & (df['英文'] < 60)]
```

## GroupBy

```python
df = pd.DataFrame({'Animal': ['Falcon', 'Falcon',
                              'Parrot', 'Parrot'],
                   'Max Speed': [380., 370., 24., 26.]})
```

```python
df
```

```python
df.groupby(['Animal']).mean()
```

```python
df.groupby(['Animal']).count()
```

```python
df.groupby(['Animal']).sum()
```

```python
groups = df.groupby(['Animal'])
```

```python
groups.get_group('Falcon')
```

```python
groups.mean()
```
