```python
from PIL import Image
import numpy as np
```

```python
im = Image.open("lin.jpg")
```

```python
im = im.resize((im.size[0]//8, im.size[1]//8))
```

```python
im
```

```python
im_array = np.array(im)
```

```python
im_array.shape
```

```python
# im_array[0, :] = (255,255,255)
im_array[0, :] = 0
```

```python
# im_array[179, :] = (255,255,255)
im_array[179, :] = 0
```

```python
# im_array[:, 0] = (255,255,255)
im_array[:, 0] = 0
```

```python
# im_array[:, 134] = (255, 255, 255)
im_array[:, 134] = 0
```

```python
new_im = Image.fromarray(im_array)
```
